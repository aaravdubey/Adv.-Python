```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = np.zeros((512,512,3), np.uint8)

cv.line(img,(0,0),(511,511),(255,0,0),5)
cv.line(img,(255,0),(255,511),(0,255,0),5)
cv.line(img,(511,0),(0,511),(0,0,255),5)

# cv.imshow("Original Image", img)
# cv.waitKey(0) 
# cv.destroyAllWindows()
plt.imshow(img)
```

```python
img = np.zeros((512,512,3), np.uint8)

cv.line(img,(100,100),(100,400),(255,0,0),5)
cv.line(img,(100,400),(400, 400),(255,0,0),5)
cv.line(img,(400, 400),(400, 100),(255,0,0),5)
cv.line(img,(400,100),(100,100),(255,0,0),5)


plt.imshow(img)
```

```python
img = np.zeros((512,512,3), np.uint8)

cv.rectangle(img, (100, 100), (400, 400), (0, 0, 255), 5) #-1 to fill

plt.imshow(img)
```

```python
img = np.zeros((512,512,3), np.uint8)

cv.circle(img, (255, 255), 200, (0, 0, 255), -1)

plt.imshow(img)
```

```python
img = np.zeros((512,512,3), np.uint8)

cv.ellipse(img,(256,256),(100,50),0,0,180,255,-1)

plt.imshow(img)
```
