```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("D:\Wallpapers\colorful.jpg")

cv2.imshow("Original Image", img)
cv2.waitKey(0) 
cv2.destroyAllWindows()

plt.plot(), plt.imshow(img), plt.title('Original Image'), plt.axis('off')
```

```python
resized = cv2.resize(img, (400, 400))

plt.plot(), plt.imshow(resized), plt.title('Resized Image'), plt.axis('off')
```

```python
grayscaled = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

plt.plot(), plt.imshow(grayscaled), plt.title('Grayscaled Image'), plt.axis('off')
```

```python
roi = grayscaled[0:200, 0:200]

plt.plot(), plt.imshow(roi), plt.title('Cropped Image'), plt.axis('off')
```

```python
(height, width) = roi.shape[:2]
center = (width / 2, height / 2)

angle = 45
scale = 1.0
matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(roi, matrix, (width, height))

plt.plot(), plt.imshow(rotated), plt.title('Rotated Image (45deg)'), plt.axis('off')
```

```python
blurred = cv2.GaussianBlur(rotated, (5,5), 0)

plt.plot(), plt.imshow(blurred), plt.title('Blurred Image'), plt.axis('off')

# For a gaussian blur of radius 2, kernel size would be 5x5 (approx 2x of radius)
# 0 is the standard deviation, which is automatically calculated
```

```python
plt.figure(figsize=(25,25))

plt.subplot(321), plt.imshow(img), plt.title('Original Image', fontsize=20), plt.axis('off')
plt.subplot(322), plt.imshow(resized), plt.title('Resized Image', fontsize=20), plt.axis('off')
plt.subplot(323), plt.imshow(grayscaled), plt.title('Grayscaled Image', fontsize=20), plt.axis('off')
plt.subplot(324), plt.imshow(roi), plt.title('Cropped Image', fontsize=20), plt.axis('off')
plt.subplot(325), plt.imshow(rotated), plt.title('Rotated Image (45deg)', fontsize=20), plt.axis('off')
plt.subplot(326), plt.imshow(blurred), plt.title('Gaussian Blurred Image', fontsize=20), plt.axis('off')

plt.savefig('Output_image')
plt.show()
```
