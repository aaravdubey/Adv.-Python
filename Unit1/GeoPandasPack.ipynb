```python
import pandas as pd
import geopandas as gpd
import matplotlib
import contextily as ctx
```

```python
world_gdf = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
world_gdf
world_gdf.crs
```

```python
world_gdf['pop_density'] = world_gdf.pop_est/world_gdf.area * 10**6
world_gdf.sort_values(ascending=False, by='pop_density')
```

```python
world_gdf.plot('pop_density', legend=True, figsize=(20, 11))
```

```python
norm = matplotlib.colors.LogNorm(vmin=world_gdf.pop_density.min(), vmax=world_gdf.pop_density.max())

world_gdf.to_crs('epsg:4326').plot("pop_density", figsize=(20,11), legend=True, norm=norm)
```
