```python
import ipympl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib widget

df = pd.read_csv("Iris.csv")
setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']
```

```python
df.head()
df.info()
df.shape
df.describe()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.view_init(elev=90, azim=280, roll=10)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.scatter(df['PetalWidthCm'], df['PetalLengthCm'], df['SepalLengthCm'])

ax.set_title('3D Scatter plot')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.plot3D(sorted(df['PetalWidthCm']), sorted(df['PetalLengthCm']), sorted(df['SepalLengthCm']))

ax.set_title('3D Scatter plot')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.bar(setosa['Id'], setosa['PetalLengthCm'], zs=1, zdir='x')
ax.bar(range(0, 50), versicolor['PetalLengthCm'], zs=2, zdir='x')
ax.bar(range(0, 50), virginica['PetalLengthCm'], zs=3, zdir='x')

ax.set_title('3D Scatter plot')
ax.set_xlabel('Species')
ax.set_ylabel('Index')
ax.set_zlabel('PetalLengthCm')
plt.show()
```
