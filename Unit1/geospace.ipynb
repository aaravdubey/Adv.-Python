```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5))
# ortho, merc, robin
m = Basemap(projection="lcc", resolution='c', width=8E6, height=8E6, lat_0=20.5937, lon_0=78.9629)
m.bluemarble(scale=0.5)
x, y = m(75, 22)
plt.plot(x, y, 'ok', markersize=5)
plt.text(x, y, "India", fontsize=12)
```

```python
plt.figure(figsize=(20, 20))

m = Basemap()
m.drawcoastlines()
m.drawcountries(linestyle='dashed', color='red')
m.drawrivers(color='aqua')
m.fillcontinents(color='green', lake_color="aqua")
m.drawmapboundary(fill_color='aqua')
m.drawparallels(np.arange(-90.,120.,30.), color='gray', linewidth=1.0, dashes=[4,4], labels=[0, 1, 1, 0])
m.drawmeridians(range(0,360,20), color='gray', linewidth=1.0, dashes=[4,4], labels=[1, 0, 0, 1])
m.drawstates(color='black', linestyle='solid', linewidth=1.0)

plt.text(73, 22, 'INDIA', fontsize=12, color="white", weight='bold')
plt.show()
```
