```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("UScereal.csv")
```

```python
data.head()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='calories', y='fat', data=data)
plt.title('Bar Chart of Calories for Fat')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='calories', y='fat', hue='vitamins', data=data)
plt.title('Bar Chart of Calories for Fat with Vitamins')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories', y='fat', data=data)
plt.title('Scatter Plot of Calories vs Fat')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='calories', y='fat', hue='vitamins', data=data)
plt.title('Scatter Plot of Calories vs Fat with Vitamins')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='calories', kde=True)
plt.title('Histogram of Calories')
plt.show()

plt.figure(figsize=(12, 8))
sns.histplot(data=data, x='calories', hue='mfr', kde=True)
plt.title('Histogram of Calories with Manufacturer')
plt.show()
```

```python
plt.figure(figsize=(8, 8), facecolor='white')
data['mfr'].value_counts().plot.pie(autopct='%1.1f%%', labels=data['mfr'].value_counts().index)
plt.title('Pie Chart of Cereal Manufacturer Distribution')
plt.show()
```

```python
sns.pairplot(data, hue='mfr', height=2.5)
plt.suptitle('Pair Plot of USCereal Dataset', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='protein', y='fibre', size='vitamins', data=data, hue='fat')
plt.title('Bubble Plot of Fibre vs Protein with fat and vitamins')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='mfr', y='protein', data=data)
plt.title('Box Plot of Protein and Manufacturers')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='shelf', y='sodium', data=data)
plt.title('Violin Plot of Sodium and Shelf')
plt.show()
```

```python
data.plot.area(stacked=False, figsize=(12, 8))
plt.title('Area Plot of Cereal Features')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=data.shelf, y=data.protein, label='Sine Curve')

plt.xlabel('Shelf')
plt.ylabel('Protein')
plt.title('Simple Line Plot of Shelf and Protein')

plt.legend()
plt.show()
```
