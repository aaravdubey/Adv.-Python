```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits import mplot3d
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
df.info()
df.shape
df.describe()
```

```python
df.iloc[:,5].unique()
```

```python
df.groupby('Species').size()
df.groupby(['Species', 'PetalWidthCm']).size()
```

```python
df = df.drop(columns="Id")
df.describe()
```

```python
setosa = df[df['Species'] == 'Iris-setosa']
versicolor = df[df['Species'] == 'Iris-versicolor']
virginica = df[df['Species'] == 'Iris-virginica']
data = setosa["SepalWidthCm"]
data.plot(kind='hist', label="setosa", color='r', fontsize=10)
plt.title("setosa")
plt.show()
```

```python
setosa.plot(x="SepalLengthCm", y="SepalWidthCm", kind='scatter', label="setosa", color='r', fontsize=10)
plt.title("setosa")
plt.show()
```

```python
data.plot(x="SepalLengthCm", kind='line', label="setosa", color='r', fontsize=10)
plt.title("setosa")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
setosa["SepalWidthCm"].plot(kind="line", color="r")
versicolor["SepalWidthCm"].plot(kind="line", color="g")
virginica["SepalWidthCm"].plot(kind="line", color="b")
# plt.axis([50, 120, 2, 5])   #Obsrevations 50 to 120
plt.legend(['setosa', 'versicolor', 'virginica'])
plt.show()
```

```python
plt.figure(figsize=(20, 5))
setosa.plot.area(stacked=False)
plt.show()
setosa.describe()
```

```python
types = df['Species'].unique()
count = df['Species'].value_counts()
avg = df.groupby('Species')['SepalWidthCm'].mean()

plt.bar(types, avg, color=['r', 'g','b'])
plt.show()
plt.barh(types, avg, color=['r', 'g','b'])
plt.show()
```

```python
plt.figure(facecolor='white')
plt.pie(avg, explode=(0.1, 0.2, 0), labels=types)
plt.legend()
plt.show()
```

```python
setosa.plot.box(notch=True,vert=False)
# plt.legend(labels)
```

```python
sns.set(style="dark")

sns.pairplot(df, hue="Species")
```

```python
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.bar(df['PetalLengthCm'], df['PetalWidthCm'], zdir='y')
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(projection='3d')

colors = ['r', 'g', 'b']
yticks = [2, 1, 0]
for c, k in zip(colors, yticks):
    xs = range(0, 50)
    ys = df['PetalLengthCm'].iloc[:50]

    cs = [c] * len(xs)

    ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_yticks(yticks)

plt.show()
```
